```python
import ast #for converting str to list

def convert(text):
    List = []
    for i in ast.literal_eval(text):
        List.append(i['name']) 
    return List
```

```python
data['genres'] = data['genres'].apply(convert)
```

```python
data.head(2)
```

```python
data.iloc[0]['keywords']
```

```python
data['keywords'] = data['keywords'].apply(convert)
data.head(2)
```

```python
data.iloc[0]['cast']
```

```python
# Chấp nhận lấy 3 diễn viên đầu tiên vì chưa có dữ liệu bổ sung

def convert_cast(text):
    List = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            List.append(i['name'])
        counter+=1
    return List
```

```python
data['cast'] = data['cast'].apply(convert_cast)
data.head(2)
```

```python
#Lấy tên đạo diễn

def convert_crew(text):
    List = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            List.append(i['name'])
            break
    return List
```

```python
data['crew'] = data['crew'].apply(convert_crew)
data.head(2)
```

```python
data.iloc[0]['overview']
```

```python
#Tách các chuỗi thành list các từ
data['overview'] = data['overview'].apply(lambda x:x.split())
data.head(2)
```

```python
#Bỏ space để các cụm từ gắn với nhau, không bị tách khi train

def remove_space(List):
    newList = []
    for i in List:
        newList.append(i.replace(" ",""))
    return newList
```

```python
data['cast'] = data['cast'].apply(remove_space)
data['crew'] = data['crew'].apply(remove_space)
data['genres'] = data['genres'].apply(remove_space)
data['keywords'] = data['keywords'].apply(remove_space)
data.head(2)
```

```python
# Nối các cột vừa xử lý lại thành cột tags
data['tags'] = data['overview'] + data['genres'] + data['keywords'] + data['cast'] + data['crew']
data.head(2)
```

```python
data.iloc[0]['tags']
```

```python
#Bỏ các cột vừa gộp
data = data[['id', 'title', 'tags', 'year', 'vote_average']]
data.head(2)
```

```python
#Chuyển list thành chuỗi
data['tags'] = data['tags'].apply(lambda x: " ".join(x))
data.head(2)
```

```python
# Chuyển tất cả về chữ thường
data['tags'] = data['tags'].apply(lambda x:x.lower())
data.head(2)
```

```python
ps = PorterStemmer()
```

```python
#Gom các từ gần giống nhau về làm một
#Ví dụ: house, houses -> house
def stems(text):
    T = []
    
    for i in text.split():
        T.append(ps.stem(i))
    
    return " ".join(T)
```

```python
data['tags'] = data['tags'].apply(stems)
```

### Sử dụng CountVectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(data['tags']).toarray()
vector
```

```python
#Tổng cộng 5000 từ được học
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
```

```python
#Hàng i: Phim thứ i
#Hàng j: Phim thứ j
#similarity[i][j]: Độ tương đồng phim i và j
similarity.shape 
```

```python
def recommend(movie):
    #Lấy ra danh sách index của phim cần tìm
    #Lưu ý có thể có nhiều index vì có các phim trùng title
    indexList = data[data['title'].str.lower() == movie.lower()]
    
    #Xử lý trường hợp phim không tồn tại
    if indexList.empty:
        print("Movie not found")
        return
        
    #Lấy phim đầu tiên trong danh sách các phim trùng title
    index = indexList.index[0]
        
    #enumerate(similarity[index]) return tuple (index_movie, score)
    #Sắp xếp thành list giảm dần theo similarity score
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1]) 

    #Lấy 5 phim giống nhất, bỏ phim đầu tiên vì là chính phim đang tìm
    for i in distances[1:11]:
        print(data.iloc[i[0]].title)
    return distances
```

```python
List = recommend("the prestige")
```
